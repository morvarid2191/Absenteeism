```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the system of equations
A = np.array([[3, 5, 2, 1], [4, 5, 0, 1], [2, 7, -1, -1]])
b = np.array([1, 3, 10])
x, residuals, rank, s = np.linalg.lstsq(A, b, rcond=None)

# Define the range of x, y, and z values to plot
x_range = np.linspace(-5, 5, 10)
y_range = np.linspace(-5, 5, 10)
X, Y = np.meshgrid(x_range, y_range)

# Solve for the corresponding z values using the equation of the plane
Z1 = (1 - 3*X - 5*Y - x[2]*X - x[3]*Y) / x[0]
Z2 = (3 - 4*X - 5*Y - x[3]*X) / x[1]
Z3 = (10 - 2*X - 7*Y + x[2]*X + x[3]*Y) / (-x[2])

# Plot the surfaces corresponding to each equation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z1, alpha=0.5)
ax.plot_surface(X, Y, Z2, alpha=0.5)
ax.plot_surface(X, Y, Z3, alpha=0.5)

# Plot the solution point
ax.scatter(x[0], x[1], x[2], color='red')

# Set the axis labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set the plot limits
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

plt.show()
```
